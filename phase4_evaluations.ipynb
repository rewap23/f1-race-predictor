```python
# testing the model on a 2025 race

import joblib
import pandas as pd

# Load saved model and feature list
rf_loaded = joblib.load("model/random_forest.pkl")
feature_cols = joblib.load("model/feature_cols.pkl")

# Load your full dataset
df = pd.read_csv("data/features.csv")

# Pick a specific 2025 race to inspect — Australian GP is Round 1
test_race = df[(df["Season"] == 2025) & (df["Round"] == 1)].copy()

# Get predictions
test_race["PodiumProbability"] = rf_loaded.predict_proba(
    test_race[feature_cols]
)[:, 1]

# Rank drivers by predicted probability
predictions = test_race[["Driver", "GridPosition", "PodiumProbability", "Podium"]].copy()
predictions = predictions.sort_values("PodiumProbability", ascending=False).reset_index(drop=True)

print("=== 2025 Australian GP — Model Predictions vs Reality ===\n")
print(predictions.to_string(index=False))

# Top 3 predicted
print("\n--- Predicted podium ---")
print(predictions.head(3)[["Driver", "PodiumProbability"]].to_string(index=False))

# Actual podium
print("\n--- Actual podium ---")
actual = predictions[predictions["Podium"] == 1][["Driver", "PodiumProbability"]]
print(actual.to_string(index=False))
```
